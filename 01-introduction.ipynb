## ICE Magic

Throughout these notebooks, we will be working with code that we need to edit, compile and execute during the exercises.
Take the [Hello World](code/introduction/hello-world.cpp) code as an example.
Left-click the link to the file to open it in a new tab in Jupyter Lab.

We could now edit this code (not necessary here), save it and then compile the code on the command line, e.g.:

```python
!g++ -O3 -o code/introduction/hello-world.out code/introduction/hello-world.cpp
```

The compiled application could now be executed on the command line with:

```python
!./code/introduction/hello-world.out
```

These steps add additional friction (opening the file, saving, compiling and running the code).
In addition, each code file contains the necessary includes, which might be distracting.

To reduce friction and remove distracting boiler platecode, we use **magic commands**.
The below cell loads an extension called **ICE**, short for instantiate-compile-execute.

```python
%load_ext ice.magic
```

After the extension is loaded, we can use the `%%cpp` magic command.
It takes the remaining contents of the cell it is included in and does the following steps:
* *Instantiate*: The code provided is augmented with auxiliary boilerplate code such as standard includes and written to file.
* *Compile*: The generated file is compiled to a binary.
* *Execute*: The binary is executed.


We start with a simple hello world example to illustrate how ICE can be used.

```python
%%cpp
  👆

int main(int argc, char *argv[]) {
    printf("Hello world from a compiled application\n");

    return 0;
}
```

This cell contains only the relevant code and omits boilerplate includes.
When running this cell, the `%cpp` magic command adds these necessary code lines, writes the code to a file, calls the compiler and executes the cell.
Thus, we can stay within the same file in Jupyter Lab during the sessions and need to run fewer commands as well.

Throughout this course, 👆 markers appear in code cells as visual annotations highlighting key parts of the code.
They are stripped out by ICE before compilation and have no effect on the generated program.

Our magic function also takes additional parameters which can be displayed with the following cell.

```python
%%cpp --help
//#
```

Let's specify the (code) output file (`-c`) as well as add the *verbose* (`-v`) and *display* (`-d`) flags.

```python
%%cpp -c code/introduction/hello-world.cpp -v -d
                                           👆 👆

int main(int argc, char *argv[]) {
    printf("Hello world from a compiled application\n");

    return 0;
}
```

We use the `-c` option for all exercises. 
The code files live in the subdirectory for the current notebook within `code/`.
From the verbose output, we can see that the ICE magic command stores the compiled files in `$HOME/app`.
